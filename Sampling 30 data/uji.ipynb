```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
import warnings
warnings.filterwarnings
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.stats.stattools import durbin_watson
from scipy.stats import t
from statistics import stdev
```

```python
df = pd.read_excel("uji.xlsx")
print(df)
```

```python
nilai_max = 5 * len(df)
```

```python
df.loc['jumlah'] = df.sum()
print(df)
```

```python
df.loc['%per item'] = (df.loc['jumlah'] / nilai_max) * 100
print(df)
```

```python
# Rata persen tiap Variabel
rata_persen_x1 = (df.loc["%per item", "X1.1":"X1.8"].sum()) / 8
rata_persen_x2 = (df.loc["%per item", "X2.1":"X2.5"].sum()) / 5
rata_persen_x3 = (df.loc["%per item", "X3.1":"X3.3"].sum()) / 3
rata_persen_x4 = (df.loc["%per item", "X4.1":"X4.2"].sum()) / 2
rata_persen_y = (df.loc["%per item", "Y1":"Y2.8"].sum()) / 2
print(round(rata_persen_x1,2))
print(round(rata_persen_x2,2))
print(round(rata_persen_x3,2))
print(round(rata_persen_x4,2))
print(round(rata_persen_y,2))
```

```python
# Mean dari jumlah % keseluruhan item
mean = df.loc["%per item"].sum()/len(df.loc["%per item"])
print(round(mean))
```

```python
sd = stdev(df.loc["%per item"])
print(round(sd))
```

```python
# Menentukan titik batas tinggi, sedang, dan rendah
a = mean-sd
b = mean+sd
print(a)
print(b)
```

```python
x = 66

if x < a:
    print("Rendah")
elif x>= b:
    print("Tinggi")
else :
    print("Sedang")
```
